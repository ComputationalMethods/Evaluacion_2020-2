```
#primero trabajaremos odeint 2 dimensiones
#importamos primero las librerias
import scipy.integrate as sint
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
m = 0.012277471 
l=1
c = (m,l)
def f1(I,t,m,l):
    f1 = l*I[1]
    f3 = I[3]
    f2 = I[0] + 2*I[3] - ((1-m)*(I[0]+m))/((I[0]+m)**2+I[2]**2)**(3/2) - (m*(I[0]-1+m))/((I[0]-1+m)**2+I[2]**2)**(3/2) 
    f4 = I[2] - 2*I[1] - ((1-m)*I[2])/((I[0]+m)**2+I[2]**2)**(3/2) - m*I[2]/((I[0]-1+m)**2+I[2]**2)**(3/2)
    return np.array([f1,f2,f3,f4])
I = [0.994, 0, 0, -2.0015851063790825224053786224] #x,vx,y,vy
T_=17.0652165601579625588917206249
t_1=np.linspace(0, 3*T_, 24000)
y_ = sint.odeint(f1,I,t_1,c)

plt.figure(figsize=(8,6))
plt.subplot(211)
plt.title('x y vx vs t')
plt.xlabel('t')
plt.ylabel('x(t)')
plt.plot(t_1,y_[:,0],'y',label='Posición')
plt.plot(t_1,y_[:,1],'m-.',label='Velocidad')
plt.legend()
plt.show()
    
```

```
## Ahora veamos Odeint 3 dimensiones
I_1 = [0.994,0,0,-2.0015851063790825224053786224,1,0]
l=1
cte = (m,l)
def f(I,t,m,l):
    f1 = l*I[1]
    f2 = I[0] + 2*I[3] - ((1-m)*(I[0]+m))/((I[0]+m)**2+I[2]**2+I[4]**2)**(3/2) - (m*(I[0]-1+m))/((I[0]-1+m)**2+I[2]**2+I[4]**2)**(3/2) 
    
    f3 = I[3]
    f4 = I[2] - 2*I[1] - ((1-m)*I[2])/((I[0]+m)**2+I[2]**2+I[4]**2)**(3/2) - m*I[2]/((I[0]-1+m)**2+I[2]**2+I[4]**2)**(3/2)
    
    f5 = I[5]
    f6 = -(I[4]*(1-m))/((I[0]+m)**2+I[2]**2+I[4]**2)**(3/2) - (I[4]*m)/((I[0]-1+m)**2+I[2]**2+I[4]**2)**(3/2)
    
    return np.array([f1,f2,f3,f4,f5,f6])

t_2 = np.linspace(0,3*T_,24000)
y2 = sint.odeint(f,I_1,t_2,cte)

plt.figure(figsize=(8,6))
plt.subplot(311)
plt.title('Posición y velocidad vs tiempo')
plt.xlabel('t')
plt.ylabel('x(t)')
plt.plot(t_2,y2[:,0],'y.-',label='Posición')
plt.plot(t_2,y2[:,1],'d',label='Velocidad')
plt.legend()
    
plt.subplot(312)
plt.xlabel('t')
plt.ylabel('y(t)')
plt.plot(t_2,y2[:,2],'m',label='Posición')
plt.plot(t_2,y2[:,3],'b',label='Velocidad')
plt.legend()
    
plt.subplot(313)
plt.xlabel('t')
plt.ylabel('z(t)')
plt.plot(t_2,y2[:,4],'g',label='Posición')
plt.plot(t_2,y2[:,5],'b',label='Velocidad')
plt.legend()
   
plt.show()
 
plt.figure(figsize=(8,6))
plt.subplot(131)
plt.xlabel('x')
plt.ylabel('dx/dt')
plt.plot(y2[:,0],y2[:,1],'b')
    
plt.subplot(132)
plt.title('Posición vs velocidad')
plt.xlabel('y')
plt.ylabel('dy/dt')
plt.plot(y2[:,2],y2[:,3],'g')

plt.subplot(133)
plt.xlabel('z')
plt.ylabel('dz/dt')
plt.plot(y2[:,4],y2[:,5],'d')

plt.show() 
```
