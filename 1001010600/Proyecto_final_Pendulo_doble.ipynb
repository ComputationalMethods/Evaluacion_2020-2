 # **Péndulo doble**

El péndulo doble es un sistema compuesto por dos péndulos, con el segundo colgando del extremo del primero. En este caso se va a estudiar un péndulo doble plano, es decir, ambos péndulos son coplanares como se muestra en la siguiente figura.
<h1><center>

![IMG](https://drive.google.com/uc?export=view&id=1G6nnWCvWANfObBVmJuMNiEQN-xsmrrVD)

</center></h1>

Donde $\theta_i$ representa el ángulo del péndulo respecto a la vertical, $l_i$ representa la longitud de la varilla y $m_i$ la masa del respectivo péndulo.

Las ecuaciones que describen la posición de las masas son:

<h1><center>

$x_1 = l_1 sin\theta_1 \qquad y_1 = -l_1 cos\theta_1$

$x_2 = l_1 sin\theta_1 + l_2 sin\theta_2 \qquad y_2 = -l_1 cos\theta_1 -l_2 cos\theta_2$

Con $\theta_1$ y $\theta_2$ dependientes del tiempo


</center></h1>

Para hallar las ecuaciones que describen el movimiento de estos péndulos se usará el Lagrangiano. Para esto hallamos la energía cinética $T$ y la energía potencial $V$.

Las velocidades de las masas son:

<h1><center>

$\vec{v_1} = l_1\dot{\theta_1}cos\theta_1 \hat{i}+ l_1\dot{\theta_1}sin\theta_1 \hat{j}$ 

$\vec{v_2} = (  l_2\dot{\theta_2}cos\theta_2 + l_1\dot{\theta_1}cos\theta_1 )\hat{i}+ (l_2\dot{\theta_2}sin\theta_2 + l_1\dot{\theta_1}sin\theta_1 )\hat{j}$

</center></h1>

Ahora hallando la energía cinética y la energía potencial, se obtiene:

<h1><center>

$T = \frac{1}{2}m_1 l_1^2(\dot{\theta_1})^2 + \frac{1}{2}m_2 l_1^2(\dot{\theta_1})^2 + \frac{1}{2}m_2 l_2^2(\dot{\theta_2})^2 + m_2l_1l_2\dot{\theta_1}\dot{\theta_2}cos(\theta_1 - \theta_2) $

$V = -m_1gl_1cos\theta_1 -m_2g(l_1cos\theta_1 + l_2\theta_2) = -(m_1 + m_2)gl_1cos\theta_1 -m_2gl_2cos\theta_2 $

</center></h1>

El lagrangiano se calcula como:

<h1><center>

$\mathscr{L} = T - V$

$\mathscr{L} = \frac{1}{2}m_1 l_1^2(\dot{\theta_1})^2 + \frac{1}{2}m_2 l_1^2(\dot{\theta_1})^2 + \frac{1}{2}m_2 l_2^2(\dot{\theta_2})^2 + m_2l_1l_2\dot{\theta_1}\dot{\theta_2}cos(\theta_1 - \theta_2) + (m_1 + m_2)gl_1cos\theta_1 + m_2gl_2cos\theta_2 $

</center></h1>

Las ecuaciones del movimiento de Lagrange nos llevan al sistema de dos ecuaciones diferenciales de segundo orden.

<h1><center>

$\frac{\partial}{\partial t}\frac{\partial \mathscr{L}}{\partial \dot{\theta_1}} - \frac{\partial \mathscr{L}}{\partial \theta_1} = 0$

$\frac{\partial}{\partial t}\frac{\partial \mathscr{L}}{\partial \dot{\theta_2}} - \frac{\partial \mathscr{L}}{\partial \theta_2} = 0$

</center></h1>

Realizando las respectivas derivadas obtenemos:

<h1><center>

$(m_1+m_2)l_1^2\ddot{\theta_1} + m_2l_1l_2\ddot{\theta_2}cos(\theta_1 - \theta_2)+m_2l_1l_2(\dot{\theta_2})^2sin(\theta_1 - \theta_2) + (m_1+m_2)gl_1sin\theta_1 = 0$

$m_2l_2^2\ddot{\theta_2} + m_2l_1l_2\ddot{\theta_1}cos(\theta_1 - \theta_2)-m_2l_1l_2(\dot{\theta_1})^2sin(\theta_1 - \theta_2) + m_2gl_2sin\theta_2 = 0$

</center></h1>

Así obtenemos un sistema de dos ecuaciones con dos incognitas, donde tenemos que despejar $\ddot{\theta_i}$



\begin{equation}
\left\lbrace
\begin{array}{ll}
a_1\ddot{\theta_1} + b\ddot{\theta_2} + c_1 = 0 \\
a_2\ddot{\theta_2} + b\ddot{\theta_1} + c_2  = 0
\end{array}
\right.
\end{equation}

Despejando obtenemos:

<h1><center>

$\ddot{\theta_1} = \frac{a_2c_1 - bc_2}{b^2-a_1a_2}$

$\ddot{\theta_2} = \frac{a_2c_1 - bc_1}{b^2-a_1a_2}$

</center></h1>

Llamamos $m = \frac{m_1}{m_2}$ y el resultado es el siguiente:

<h1><center>

$\ddot{\theta_1} = \frac{cos(\theta_1 - \theta_2)(\frac{g}{l_1}sin\theta_2-(\dot{\theta_1})^2sin(\theta_1-\theta_2))-\frac{l_2}{l_1}((m+1)\frac{g}{l_2}sin\theta_1+(\dot{\theta_2})^2sin(\theta_1-\theta_2))}{m+sin^2(\theta_1-\theta_2)}$

$\ddot{\theta_2} = \frac{cos(\theta_1 - \theta_2)((m+1)\frac{g}{l_2}sin\theta_1+(\dot{\theta_2})^2sin(\theta_1-\theta_2))-(m+1)\frac{l_1}{l_2}(\frac{g}{l_1}sin\theta_2-(\dot{\theta_1})^2sin(\theta_1-\theta_2))}{m+sin^2(\theta_1-\theta_2)}$


</center></h1>



## Solución numérica

Para hallar la solución numérica se implentaran metodos de Scipy.


```
import numpy as np
from scipy import integrate


#Ecuaciones diferenciales a integrar

def ang_vel(initial_conditions, t, parametros):
  '''
  Aqui se definen las ecuaciones diferenciales para el pendulo doble
  '''

  m1, m2, l1, l2, g = parametros
  a1, a2, w1, w2 = initial_conditions

  #Definimos los terminos que componen las ecuaciones diferenciales

  #Masa
  m = m1/m2

  #Denominador
  denom = m + np.sin(a1-a2)**2

  #velocidad angular 1
  term1 = np.cos(a1-a2)*( (g/l1)*np.sin(a2) - w1**2 * np.sin(a1-a2) )
  term2 = (l2/l1)*( (m+1)*(g/l2)*np.sin(a1) + w2**2 * np.sin(a1-a2) )

  #velocidad angular 2

  term3 = np.cos(a1-a2)*( (m+1)*(g/l2)*np.sin(a1) + w2**2 * np.sin(a1-a2) )
  term4 = (m+1)*(l1/l2)*( (g/l1)*np.sin(a2) - w1**2 * np.sin(a1-a2) )

  #Se define un vector con las ecuaciones respectivas
  
  A = [w1, w2,(term1 - term2)/denom, (term3 - term4)/denom ]

  return A

#Parametros 

#Masas
m1 = 1
m2 = 1
#Longitud de las varillas
l1 = 1
l2 = 1
#Aceleracion de la gravedad
g = 9.8

#Condiciones iniciales

#Angulos iniciales
a1 = np.pi/2
a2 = np.pi/2
#Velocidad angular inicial
w1 = 0
w2 = 0

parametros =[m1, m2, l1, l2, g] #Parametros del sistema
initial_conditions = [a1, a2, w1, w2] #Condiciones iniciales del sistema
t = np.linspace(0, 1000, 20000) #Candtidad de puntos para hallar la solucion

#Hallando las soluciones a las ecuaciones diferenciales
solutions = integrate.odeint(ang_vel, initial_conditions, t, args = (parametros, ) ) 
```

Una vez se obtiene la solución, se procede a graficar las posiciones de las masas.

```
import matplotlib.pyplot as plt

theta1 = solutions[:,0]
theta2 = solutions[:,1]

#Posiciones
x1 = l1*np.sin(theta1)
y1 = -l1*np.cos(theta1)

x2 = l1*np.sin(theta1) + l2*np.sin(theta2)
y2= -l1*np.cos(theta1) - l2*np.cos(theta2)

plt.plot(x1[::], y1[::], 'bo', ms = 0.1)
plt.plot(x2[::], y2[::], 'ro', ms = 0.1)
#plt.xlim(-1,1)
#plt.ylim(-1,1)
plt.show()
```

#Animación
Para visualizar mejor el fenómeno, se realizó una animación usando la librería pygame. La animación se puede ver en [Animación](https://drive.google.com/file/d/124GKcH8J9wdopxlwu73vkA_deDj0V-4o/view?usp=sharing)

El código usado para realizar esta animación fue el siguiente:

```
#Simulacion


width, height = 1080,720
size =(width, height)
pygame.init()
pygame.display.set_caption('Pendulo doble')
fps = 30
screen = pygame.display.set_mode(size)
clock = pygame.time.Clock()


starting_point = (int(width/2),int(height/4))

run = True

white = (255,255,255)
black = (0,0,0)
i=0 #Iterador

#Lista para la Trayectoria
tray1 = []
tray2 = []

while run:
    clock.tick(fps)
    screen.fill(black)
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            run = False

    pos1 = x1[i]+starting_point[0], y1[i]+starting_point[1]
    pos2 = x2[i]+starting_point[0], y2[i]+starting_point[1]

#Movimiento del pendulo
    pygame.draw.line(screen, white, starting_point, (pos1), 2)
    pygame.draw.line(screen, white,(pos1), (pos2), 2)

    pygame.draw.circle(screen, (17,208,238),  (pos1), 8)
    pygame.draw.circle(screen, (201,17,238), (pos2), 8)

#Trayectoria del Pendulo

    tray1.append((pos1))
    tray2.append((pos2))



    if len(tray1)>1:
        pygame.draw.lines(screen, (231,1,58), False, tray1, 1)

    if len(tray2)>1:
        pygame.draw.lines(screen, (245,91,10), False, tray2, 1)



    i += 1

    pygame.display.update()

pygame.quit()
```
